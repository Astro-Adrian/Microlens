```python
# Import basic packages
import MulensModel
import matplotlib.pyplot as pl# MulensModel uses matplotlib for plotting.
import os
```

```python
(t_min, t_max) = (2457895,2457940)

# Define a point lens model:
pspl_model = MulensModel.Model(
    {'t_0': 2457916.003989545309, 'u_0': 0.020222720774864797, 't_E': 108.32254745787982})

# model with planet:
planet_model = MulensModel.Model(
    {'t_0': 2457916.243, 'u_0': 0.0303, 't_E':101.0, 'rho': 0.00664, 
     'q': 0.0015, 's': 1.0318, 'alpha': -212.200649})
# Since rho is set, define a time range and method to apply 
# finite source effects:
planet_model.set_magnification_methods(
    [2457910., 'VBBL', 2457930.])
```

```python
# Plot those models:
pspl_model.plot_magnification(
    t_range=[2457895, 2457940], subtract_2450000=True, color='red', 
    linestyle=':', label='PSPL')
planet_model.plot_magnification(
     t_range=[2457895, 2457940], subtract_2450000=True, color='black', 
     label='1S2L')
pl.legend(loc='best')
pl.show()
```
